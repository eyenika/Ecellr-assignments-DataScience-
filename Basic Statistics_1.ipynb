```python
#visualisation of data
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=pd.read_csv('sales_data_with_discounts.csv')
```

```python
numerical_columns=df.select_dtypes(include=['int','float']).columns
for columns in numerical_columns:
    plt.figure(figsize=(8,6))
    sns.histplot(df[columns],kde=True)
    plt.title(f'histogram of {columns}')
    plt.xlabel(columns)
    plt.ylabel('frequency')
    plt.show()
```

```python
categorical_columns=df.select_dtypes(include=['object']).columns
for columns in categorical_columns:
    plt.figure(figsize=(8,6))
    sns.countplot(data=df,x=columns)
    plt.title(f'bar chart of {columns}')
    plt.xlabel(columns)
    plt.ylabel('count')
    plt.show()
              
```

```python
#standardisation of numerical colums
from sklearn.preprocessing import StandardScaler
import pandas as pd
df=pd.read_csv('sales_data_with_discounts.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.title('Before Standardization')
for column in numerical_columns:
    sns.histplot(df[column], kde=True, label=column)
plt.legend()
plt.show()
```

```python
#Standardization
scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
plt.figure(figsize=(10, 6))
plt.title('After Standardization')
for column in numerical_columns:
    sns.histplot(df[column], kde=True, label=column)
plt.show()
```

```python
#conversion_of_categorical_data_to_dummyvariables
import pandas as pd
df=pd.read_csv('sales_data_with_discounts.csv')
```

```python
categorical_columns=df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
#one-hot-encoding
df_encoded=pd.get_dummies(df,columns=categorical_columns,drop_first=True)
df_encoded
```

```python
print('transformed dataset: ')
print(df_encoded.head())
```
